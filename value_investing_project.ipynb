```python
import yfinance as yf
from stock_class import Stock
from funcs import file_read
tickers = file_read('tickers.csv')

new_list = sorted(tickers, key=lambda x: x.rank_score, reverse=True)

for i in new_list:
    print("Stock Ticker - ", i.ticker, " score is  ", i.rank_score)
```

```python
import pandas as pd

rows = []

for stock in new_list:
    rows.append({
        "Ticker": stock.ticker,
        "Rank Score": stock.rank_score,
        "Revenue_to_EV_%": stock.rev_toEV,
        "Profit_Margin_%": stock.profit_margin,
        "Return_on_EV_%": stock.return_on_EV,
        "2yr_Revenue_Growth_%": stock.twoyr_growth,
        "Enterprise_Value": stock.get_enterprise_value(),
        "Market_Cap": stock.market_cap,
        "Eligible": stock.eligible
    })

df = pd.DataFrame(rows)

df.to_csv("value_rankings.csv", index=False)
df.head(10)
```
